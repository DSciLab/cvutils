```python
from cvutils import imread, imshow
from cvutils import transform as tf
```

```python
image_path = '../asset/rick.png'

image = imread(image_path)
imshow(image)
```

```python
resize = tf.Resize()
transpose_to_torch = tf.TransposeTorch()
untranspose_to_torch = tf.UntransposeTorch()

image = transpose_to_torch(image)
resized_image = resize(image, scale=0.5)
resized_image = untranspose_to_torch(resized_image)

print(resized_image.shape)
imshow(resized_image)
```

```python
resize = tf.Resize()
transpose_to_torch = tf.TransposeTorch()
untranspose_to_torch = tf.UntransposeTorch()

image = transpose_to_torch(image)
resized_image = resize(image, size=(100, 200))
resized_image = transpose_to_torch(resized_image)

print(resized_image.shape)
imshow(resized_image)
```
