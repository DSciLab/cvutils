```python
from cvutils import imread, imshow
from cvutils import transform as tf
```

```python
image_path = '../asset/rick.png'

image = imread(image_path)
imshow(image)
```

```python
flip_x = tf.FlipX()
transpose_to_torch = tf.TransposeTorch()
untranspose_to_torch = tf.UntransposeTorch()

image = transpose_to_torch(image)
fliped_x_image = flip_x(image)
fliped_x_image = untranspose_to_torch(fliped_x_image)

imshow(fliped_x_image)
```

```python
flip_y = tf.FlipY()
transpose_to_torch = tf.TransposeTorch()
untranspose_to_torch = tf.UntransposeTorch()

image = transpose_to_torch(image)
fliped_y_image = flip_y(image)
fliped_y_image = untranspose_to_torch(fliped_y_image)

imshow(fliped_y_image)
```
