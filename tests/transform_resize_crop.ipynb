```python
from cvutils import imread, imshow
from cvutils import transform as tf
```

```python
image_path = '../asset/rick.png'

image = imread(image_path)
imshow(image)
```

```python
fn = tf.ResizeRandomCrop(500, 300)
transpose_to_torch = tf.TransposeTorch()
untranspose_to_torch = tf.UntransposeTorch()
linear_norm = tf.LinearNormalize()

image = transpose_to_torch(image)
_image = linear_norm(image)
_image = fn(_image)
_image = untranspose_to_torch(_image)

imshow(_image)
```

```python
fn = tf.ResizeRandomCenterCrop(500, 300)
transpose_to_torch = tf.TransposeTorch()
untranspose_to_torch = tf.UntransposeTorch()
linear_norm = tf.LinearNormalize()

image = transpose_to_torch(image)
image = linear_norm(image)
_image = fn(image)
_image = untranspose_to_torch(_image)

imshow(_image)
```
